```python
import cv2 #Opencv kütüpahenesidir.
from keras.models import load_model #Oluşturulan cnn modelini projeye dahil etmek için
import numpy as np
from firebase import firebase #Firebase veritabanını pyhton kullanabilmek için
import json
```

```python
model = load_model('models/model_generator.h5')#Model çağırılıyor
firebase = firebase.FirebaseApplication('https://nodemculed-f73f0.firebaseio.com/', None)#Veritabanınan host adresine erişim sağlıyoruz
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
```

```python
sure=0
labels_dict={0:'Open',1:'Close'}
color_dict={0:(0,255,0),1:(0,0,255)} # 1 yani açık ise yeşil, 0 yani kapalı ise kırmızı değerine erişiyoruz

size = 4
webcam = cv2.VideoCapture(0) #Kamerayı seçiyoruz
classifier = cv2.CascadeClassifier('hcf\haarcascade_frontalface.xml')# Görüntüdeki yüzü algılamak için, haar cascade sınıflandırıcı kullanacağız.
score=0
while True:
    (rval, image) = webcam.read()#Görüntü frameler şeklinde okumay başladı.
    image=cv2.flip(image,1)#.Görüntüyü yatay eksende zıt yönde çevrilmektedir.
    image_small = cv2.resize(image, (image.shape[1] // size, image.shape[0] // size))# Görüntüyü küçülterek matrisin boyutunu da küçültüyoruz.
    faces = classifier.detectMultiScale(image_small) #Bu fonksiyon algılanan nesnenin sınırlarını yani width, height, x, y koordinatları ile ilgili bir dizi döndürür.

    for f in faces:
        (x, y, w, h) = [v * size for v in f] #Görüntüde algılanan yüzün  koordinat değerlerini değişkenlere kaydediyoruz 
        roi= image[y:y+h, x:x+w] #Sadece algılanan nesnenin bulunduğu bölgeyi kaydediyoruz.
        resized=cv2.resize(roi,(100,100)) #Modelde eğitilecek boyuta döndürüyoruz.
        normalize=resized/255.0 #Matris değerlerimizi 0-1 arasında değerlere çekiyoruz.
        reshaped=np.reshape(normalize,(1,100,100,3)) #Modelin input shape için uygun boyuta getirildi
        reshaped = np.vstack([reshaped]) #Matris yığın haline getirilmektedir.
        sinif=model.predict(reshaped) #Elde edilen nesneyi sınıflandırması için modele gönderildi
        print(sinif[0][0])
        sonuc=sinif[0][0]#Modelin tahmin sonucunu kaydediyoruz
        etiket=0
        deger=str(sonuc).find('-')
        if deger!=-1:
            etiket=1
            sure=sure+1
        else:
            etiket=0
            sure=0
        cv2.rectangle(image,(x,y),(x+w,y+h),color_dict[etiket],2)
        cv2.rectangle(image,(x,y-40),(x+w,y),color_dict[etiket],-1)
        cv2.putText(image, labels_dict[etiket]+str(sure), (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.7,(255,255,255),2)       
        if(sure>20):
            firebase.put('/RandomVal/','Value',"Close")
    cv2.imshow('LIVE',   image)
    key = cv2.waitKey(20)
    
    if key == 27: #ESC
        break
webcam.release()

cv2.destroyAllWindows()
```
